# Extracting Data

```python
import os
import pandas as pd
import json
```

```python
path_to_json ='../data/Original' 
json_files = []

for root, dirs, files in os.walk(path_to_json):
    for f in files:
        if f.endswith('.json'):      #Check for .json exten
            json_files.append(os.path.join(root, f))    #append full path to file
len(json_files)
```

```python
#key_to_extract = {'spotName', 'CurrentCurve',"VoltageCurve"}
data_list=[]
#ciclo che carica il file json specificato dal path
for i,file in enumerate(json_files):
    json_data=open(file)
    data=json.load(json_data)
    #creo ora dizionario con i soli campi di interesse
    data_list.append([data["Message"]["WeldLog"]["spotName"],data["Message"]["WeldLog"]["VoltageCurve"],
                    data["Message"]["WeldLog"]["CurrentCurve"], data["TimeStamp"]])
```

```python
#dopo aver verificato l'ordine delle feature aggiusto columns se necessario
df = pd.DataFrame (data_list, columns = ['spotName',"VoltageCurve",'CurrentCurve',"TimeStamp"])
df
```

```python
#converto TimeStamp in formato datetime
df["TimeStamp"]=pd.to_datetime(df["TimeStamp"], format='%Y-%m-%dT%H:%M:%S')
df
```

```python
#salvo df in formato pickle
df.to_pickle("../data/full_rows_df.pkl")
```
